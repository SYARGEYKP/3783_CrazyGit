```python
house = pd.read_csv('/content/kc_house_data.csv',encoding = '1251')
house.head()
```

```python
house.describe()
```

```python
chart_1=house.groupby('bedrooms').agg({'price': 'mean'}).round(0)
chart_1
```

```python
plt.figure(figsize=(8,7))
plt.hist(chart_1['price'], bins=33)
plt.xlabel('Количество спален')
plt.ylabel('Средняя стоимость')
plt.title('Распределение средней стоимости домов в зависимости от спален')
plt.xticks([i for i in range(0, 33)], rotation=45)
plt.grid();
```

Стоимость домов не зависит линейно от количества спален, есть другие факторы, влияющие на стоимость

```python
house.columns
```

```python
chart_2=house.groupby('bedrooms').agg({'sqft_living15': 'mean'}).round(0)
chart_2
```

```python
plt.figure(figsize=(8,4))
plt.hist(chart_2['sqft_living15'], bins=33)
plt.xlabel('Количество спален')
plt.ylabel('Средняя жилая площадь')
plt.title('Распределение жилой площади в зависимости от количества спален')
plt.xticks([i for i in range(0, 33)], rotation=45)
plt.grid();
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x=house['yr_built'], y =house['yr_built'].value_counts())
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.title('Распределение домов по году постройки')
plt.xticks([i for i in range(0, 117,3)], rotation=45)
```

```python
house_1 = house['waterfront'].value_counts()
house_1
```

```python
plt.pie(house_1.values, labels=house_1.index, autopct='%.1f%%')
plt.title('Распределение домов по наличию вида на набережную, 0 - нет вида, 1 есть вид');
plt.legend(house_1.index)
```

Вывод - Большая часть домов не имеет вида на набережную

```python
house_3 = house['floors'].value_counts()
house_3
```

```python
plt.figure(figsize=(15,8))
plt.pie(house_3.values, labels=house_3.index, autopct='%.1f%%')
plt.title('Распределение домов по этажности');
plt.legend(house_3.index)
```

```python
house_4 = house['condition'].value_counts()
house_4
```

```python
plt.figure(figsize=(12,8))
plt.pie(house_4.values, labels=house_4.index, autopct='%.1f%%')
plt.title('Распределение домов по состоянию');
plt.legend(house_4.index)
```

```python

from google.colab import autoviz
df_4527370754108404184 = autoviz.get_df('df_4527370754108404184')

def value_plot(df, y, sort_ascending=False, figsize=(2, 1)):
  from matplotlib import pyplot as plt
  if sort_ascending:
    df = df.sort_values(y).reset_index(drop=True)
  _, ax = plt.subplots(figsize=figsize)
  df[y].plot(kind='line')
  plt.title(y)
  ax.spines[['top', 'right',]].set_visible(False)
  plt.tight_layout()
  return autoviz.MplChart.from_current_mpl_state()

chart = value_plot(df_4527370754108404184, *['price'], **{})
chart
```
